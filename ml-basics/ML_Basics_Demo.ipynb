```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Data Loading
df = pd.read_csv("https://raw.githubusercontent.com/nachikethmurthy/Source-Code-Dataset-for-Machine-Learning-using-Python/refs/heads/main/Data/Advertising.csv")
df.head() # display first 5 rows
```

```python
# Preprocessing
df = df.drop(columns="Unnamed: 0")
df.head()
```

```python
X = df.drop(columns="sales")
y = df["sales"] # separate X and Y from dataset
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
len(X)
```

```python
len(X_train), len(X_test)
```

```python
lr = LinearRegression()
lr.fit(X_train,y_train) # training step
```

```python
lr
```

```python
lr.predict(X_test)
```

```python
y_test # existing output
```

```python
y_pred = lr.predict(X_test)
```

```python
# evaluation function
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
mean_squared_error(y_test, y_pred)
```

```python
r2_score(y_test, y_pred)
```

```python
mean_absolute_error(y_test, y_pred)
```
